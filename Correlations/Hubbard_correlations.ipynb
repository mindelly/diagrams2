```python
from scipy.fft import fft, fftfreq
```

```python
def run_fft(data: list, N: int, T:int):
    yf = fft(data)
    xf = fftfreq(N, T)[:N//2]
    plt.plot(xf, 2.0/N * np.abs(yf)[0:N//2], 'o-')
    return xf, yf, 2.0/N * np.abs(yf)[0:N//2]
```
