# Подгонка Скейлинга N/L = 1/20

```python
import lmfit
from lmfit.models import LinearModel, QuadraticModel, ExpressionModel

mod_mu = ExpressionModel('a*x+b')
mod_h = ExpressionModel('a*x+b')

# Устанавливаем начальные значения параметров (с которых стартует оптимизация)
mod_mu.set_param_hint("a", value=1.0, vary=True)
mod_mu.set_param_hint("b", value=1.0, vary=True)
mod_h.set_param_hint("a", value=1.0, vary=True)
mod_h.set_param_hint("b", value=1.0, vary=True)

_1_L = [1/60,1/80, 1/120]
mu = [mu60_2, mu80_2, mu120_2]
h = [h60_2, h80_2, h120_2]

params_mu = mod_mu.make_params()
params_h = mod_h.make_params()

result_mu = mod_mu.fit(mu, params_mu, method="leastsq", x=_1_L)  # fitting
result_h = mod_h.fit(h, params_h, method="leastsq", x=_1_L)  # fitting

print('result for h=',result_h.best_values['b'])
print('result for mu=',result_mu.best_values['b'])
```
