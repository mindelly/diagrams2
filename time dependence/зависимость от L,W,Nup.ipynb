```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
kwds = dict(ls='--', marker='o', lw=2, markersize=7)
```

# Зависимость от L

## W= 1,2,3

```python
diff_L = pd.read_csv('various_L.csv')
```

### Nup = 5, Ndown = 0

```python
diff_L_w1 = diff_L[(diff_L.W==1)&(diff_L.Nup==5)].sort_values('L').copy()
diff_L_w2 = diff_L[(diff_L.W==2)&(diff_L.Nup==5)].sort_values('L').copy()
diff_L_w3 = diff_L[(diff_L.W==3)&(diff_L.Nup==5)].sort_values('L').copy()
```

```python
plt.figure(figsize=(12, 7))
plt.title('W=1, Nup=5, Ndown=0')
plt.plot(diff_L_w1.L, diff_L_w1['time(ms)'], **kwds)
plt.xlabel("L", fontsize=20)
plt.ylabel("ms", fontsize=20)
plt.show()

plt.figure(figsize=(12, 7))
plt.title('W=2, Nup=5, Ndown=0')
plt.plot(diff_L_w2.L, diff_L_w2['time(ms)'], **kwds)
plt.xlabel("L", fontsize=20)
plt.ylabel("ms",fontsize=20)
plt.show()

plt.figure(figsize=(12, 7))
plt.title('W=3, Nup=5, Ndown=0')
plt.plot(diff_L_w3.L, diff_L_w3['time(ms)'], **kwds)
plt.xlabel("L", fontsize=20)
plt.ylabel("ms",fontsize=20)
plt.show()
```

### Nup = 6, Ndown = 1

```python
diff_L_w1 = diff_L[(diff_L.W==1)&(diff_L.Nup==6)].sort_values('L').copy()
diff_L_w2 = diff_L[(diff_L.W==2)&(diff_L.Nup==6)].sort_values('L').copy()
```
